```python
import pandas as pd
```

```python
# read raw data and merge them
data_value_path = '../data/raw/Training_Set_Values.csv'
data_label_path = '../data/raw/Training_Set_Labels.csv'
df1 = pd.read_csv(data_value_path)
df2 = pd.read_csv(data_label_path)
merged_pump_data = pd.merge(df1, df2, on="id", how="inner")
merged_pump_data.to_csv('../data/processed/pump_dataset.csv', index = False)
```
