```python
import pandas as pd
# save filepath to variable for easier access
pump_file_path = '../data/processed/df_encoded_new_features.csv'
#pump_file_path = "../data/processed/pump_dataset_full_cleaned.csv"
# read the data and store data in DataFrame titled pomp_data
pump_data = pd.read_csv(pump_file_path) 
# print a summary of the data in water pomp data
pump_data.describe()
```

```python
print(pump_data.dtypes)
```

```python
from sklearn.model_selection import train_test_split
# Select subset of predictors
cols_to_use = ['latitude', 'longitude', 'pump_age']
X = pump_data[cols_to_use]

# Select target
y = pump_data.status_group

# Separate data into training and validation sets
X_train, X_valid, y_train, y_valid = train_test_split(X, y)
```

```python
from xgboost import XGBRegressor

my_model = XGBRegressor()
my_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error

predictions = my_model.predict(X_valid)
print("Mean Absolute Error: " + str(mean_absolute_error(predictions, y_valid)))
```

```python
my_model = XGBRegressor(n_estimators=500)
my_model.fit(X_train, y_train)
```

```python
from xgboost import XGBRegressor
import xgboost
my_model = XGBRegressor(n_estimators=500)
my_model.fit(X_train, y_train, 
             regressor__early_stopping_rounds=5,
             regressor__eval_set=[(X_valid, y_valid)])
```

```python
my_model = XGBRegressor(n_estimators=1000, learning_rate=0.05)
my_model.fit(X_train, y_train, 
             early_stopping_rounds=5, 
             eval_set=[(X_valid, y_valid)], 
             verbose=False)
```

```python
my_model = XGBRegressor(n_estimators=1000, learning_rate=0.05, n_jobs=4)
my_model.fit(X_train, y_train, 
             early_stopping_rounds=5, 
             eval_set=[(X_valid, y_valid)], 
             verbose=False)
```
